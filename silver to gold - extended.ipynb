```python
import re
from pyspark.sql import functions as F

# Function to rename columns to snake_case
def rename_columns_to_snake_case(df):
    def camel_to_snake(name):
        s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
        return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1).lower()

    for col_name in df.columns:
        df = df.withColumnRenamed(col_name, camel_to_snake(col_name))
    return df

# Function to rename duplicate columns
def rename_duplicate_columns(df):
    cols = df.columns
    seen = {}
    new_cols = []

    for col in cols:
        new_col = col
        count = seen.get(col, 0)
        if count > 0:
            new_col = f"{col}_duplicate_{count}"
        seen[col] = count + 1
        new_cols.append(new_col)

    return df.toDF(*new_cols)

# Paths to the Customer and Product tables
customer_path = '/mnt/silver/SalesLT/Customer'
product_path = '/mnt/silver/SalesLT/Product'

# Read the Customer and Product Delta tables
customer_df = spark.read.format('delta').load(customer_path)
product_df = spark.read.format('delta').load(product_path)

# Apply snake_case to columns
customer_df = rename_columns_to_snake_case(customer_df)
product_df = rename_columns_to_snake_case(product_df)

# Rename duplicate columns
customer_df = rename_duplicate_columns(customer_df)
product_df = rename_duplicate_columns(product_df)

# Define the output paths
customer_output_path = '/mnt/gold/SalesLT/Customer'
product_output_path = '/mnt/gold/SalesLT/Product'

# Write the transformed DataFrames back to Delta tables (overwrite existing data)
customer_df.write.format('delta') \
    .mode('overwrite') \
    .option('mergeSchema', 'true') \
    .option('overwriteSchema', 'true') \
    .save(customer_output_path)

product_df.write.format('delta') \
    .mode('overwrite') \
    .option('mergeSchema', 'true') \
    .option('overwriteSchema', 'true') \
    .save(product_output_path)

# -----------------------------------------------
# New part to deal with the Sales table creation
# -----------------------------------------------

# Load paths for SalesOrderDetail and SalesOrderHeader
order_details_path = '/mnt/silver/SalesLT/SalesOrderDetail'
order_header_path = '/mnt/silver/SalesLT/SalesOrderHeader'

# Read the Delta tables
order_details_df = spark.read.format('delta').load(order_details_path)
order_header_df = spark.read.format('delta').load(order_header_path)

# Rename columns to snake_case
order_details_df = rename_columns_to_snake_case(order_details_df)
order_header_df = rename_columns_to_snake_case(order_header_df)

# Rename duplicate columns
order_details_df = rename_duplicate_columns(order_details_df)
order_header_df = rename_duplicate_columns(order_header_df)

# Rename 'modified_date' columns to avoid conflict
order_details_df = order_details_df.withColumnRenamed('modified_date', 'detail_modified_date')
order_header_df = order_header_df.withColumnRenamed('modified_date', 'header_modified_date')

# Join the two DataFrames on 'sales_order_id'
combined_df = order_details_df.join(
    order_header_df,
    on='sales_order_id',
    how='inner'
)

# Select only the necessary columns for the Sales table
sales_df = combined_df.select(
    'sales_order_id', 
    'sales_order_detail_id', 
    'order_qty', 
    'product_id', 
    'customer_id',
    'unit_price', 
    'unit_price_discount', 
    'line_total', 
    'detail_modified_date',
    'account_number', 
    'sales_order_number', 
    'purchase_order_number', 
    'order_date', 
    'ship_date', 
    'status', 
    'revision_number', 
    'due_date', 
    'online_order_flag', 
    'ship_to_address_id', 
    'bill_to_address_id', 
    'ship_method', 
    'sub_total', 
    'tax_amt', 
    'freight', 
    'total_due', 
    'header_modified_date'

)

# Define output path for the Sales table
sales_output_path = '/mnt/gold/SalesLT/Sales'

# Write the updated DataFrame back to the same location, enabling schema evolution
sales_df.write.format('delta') \
    .mode('overwrite') \
    .option('mergeSchema', 'true') \
    .option('overwriteSchema', 'true') \
    .save(sales_output_path)

print("Successfully applied snake_case to the Sales Delta table without creating new files.")
```
